```python
import pandas as pd

# Step 1: Load the data
df = pd.read_csv('CityTempsLatLng.csv')
```

```python
df.head()
```

```python
# Group the DataFrame by City and calculate the maximum average temperature for each city
max_temps = df.groupby('City')['AvgTemperature'].max().reset_index()

# Merge the maximum temperature DataFrame with the original DataFrame to get the lat, lng, date, and other columns
max_temps_df = pd.merge(max_temps, df, on=['City', 'AvgTemperature'], how='left')
```

```python
max_temps_df
```

```python
max_temps_df.to_csv('max_temps.csv', index=False)
```
